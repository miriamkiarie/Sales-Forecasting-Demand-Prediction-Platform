## Step 3: Utility Helpers

```python
def _find_col(df, candidates, required=True):
    """Find first matching column from a list of candidates (case-insensitive)."""
    df_cols_lower = {c.lower(): c for c in df.columns}
    for cand in candidates:
        if cand.lower() in df_cols_lower:
            return df_cols_lower[cand.lower()]
    if required:
        raise ValueError(f'Could not find any of {candidates} in: {list(df.columns)}')
    return None


def _find_file(raw_dir, candidates):
    """Find first existing file from a list of candidate filenames."""
    for fname in candidates:
        p = raw_dir / fname
        if p.exists():
            return p
    raise FileNotFoundError(
        f'None of these files found in {raw_dir}:\n  ' + '\n  '.join(candidates)
    )


def _to_standard(df, date_col, sales_col, category_col=None,
                 region_col=None, customer_col=None, profit_col=None):
    """
    Convert any dataframe to standard format:
    Order Date | Sales | Profit | Category | Region | Customer ID
    """
    out = pd.DataFrame()
    out['Order Date'] = pd.to_datetime(df[date_col], errors='coerce')
    out['Sales']      = pd.to_numeric(df[sales_col], errors='coerce').fillna(0)
    out['Profit']     = pd.to_numeric(df[profit_col], errors='coerce').fillna(0) \
                        if profit_col and profit_col in df.columns else 0
    out['Category']   = df[category_col].astype(str) \
                        if category_col and category_col in df.columns else 'General'
    out['Region']     = df[region_col].astype(str) \
                        if region_col and region_col in df.columns else 'Unknown'
    out['Customer ID']= df[customer_col].astype(str) \
                        if customer_col and customer_col in df.columns else 'Unknown'
    out = out.dropna(subset=['Order Date', 'Sales'])
    out = out[out['Sales'] > 0]
    return out


def _aggregate(df_standard):
    """Aggregate standard df to monthly series and category breakdown."""
    df_standard['Order Date'] = pd.to_datetime(df_standard['Order Date'])
    monthly = (
        df_standard.set_index('Order Date')
                   .resample('MS')['Sales']
                   .sum()
                   .rename('Total Sales')
    )
    monthly = monthly[monthly > 0]
    cat_monthly = (
        df_standard.groupby(['Order Date', 'Category'])['Sales']
                   .sum()
                   .unstack('Category')
                   .resample('MS')
                   .sum()
                   .fillna(0)
    )
    return monthly, cat_monthly


print('Helper functions defined.')
```

## Step 4: Individual Dataset Loaders

```python
def _load_superstore(raw_dir):
    for fname in ['Sample - Superstore.csv', 'superstore.csv', 'Superstore.csv']:
        p = raw_dir / fname
        if p.exists():
            df = pd.read_csv(p, encoding='latin-1')
            return _to_standard(df, 'Order Date', 'Sales', 'Category',
                                'Region', 'Customer ID', 'Profit')
    raise FileNotFoundError(
        'Superstore CSV not found.\nDownload: kaggle.com/datasets/vivek468/superstore-dataset-final\n'
        'Save as: data/raw/Sample - Superstore.csv'
    )


def _load_fmcg_profit(raw_dir):
    p = raw_dir / 'fmcg_sales_profit.csv'
    if not p.exists():
        raise FileNotFoundError('Save file as: data/raw/fmcg_sales_profit.csv')
    df = pd.read_csv(p, encoding='latin-1')
    date_col     = next((c for c in df.columns if 'date' in c.lower()), df.columns[0])
    sales_col    = next((c for c in df.columns if 'sales' in c.lower() or 'revenue' in c.lower()), None)
    profit_col   = next((c for c in df.columns if 'profit' in c.lower()), None)
    category_col = next((c for c in df.columns if 'category' in c.lower() or 'product' in c.lower()), None)
    region_col   = next((c for c in df.columns if 'region' in c.lower() or 'location' in c.lower()), None)
    if not sales_col:
        raise ValueError(f'Cannot find sales column. Available: {list(df.columns)}')
    return _to_standard(df, date_col, sales_col, category_col, region_col, profit_col=profit_col)


def _load_retail_store(raw_dir):
    train_p = raw_dir / 'train.csv'
    if not train_p.exists():
        raise FileNotFoundError('Save train.csv, stores.csv, features.csv to data/raw/')
    train = pd.read_csv(raw_dir / 'train.csv')
    train['Date'] = pd.to_datetime(train['Date'])
    if (raw_dir / 'stores.csv').exists():
        stores = pd.read_csv(raw_dir / 'stores.csv')
        train  = train.merge(stores, on='Store', how='left')
    if (raw_dir / 'features.csv').exists():
        feats = pd.read_csv(raw_dir / 'features.csv')
        feats['Date'] = pd.to_datetime(feats['Date'])
        train = train.merge(feats, on=['Store', 'Date', 'IsHoliday'], how='left')
    train['Category'] = train['Type'] if 'Type' in train.columns else 'Store'
    train['Region']   = 'Store ' + train['Store'].astype(str)
    return _to_standard(train, 'Date', 'Weekly_Sales', 'Category', 'Region')


def _load_fmcg_demand(raw_dir):
    p = _find_file(raw_dir, ['fmcg_demand.csv', 'FMCG_demand.csv'])
    df = pd.read_csv(p, encoding='latin-1')
    return _to_standard(df,
        _find_col(df, ['date', 'Date', 'order_date', 'Day']),
        _find_col(df, ['sales', 'Sales', 'demand', 'Demand', 'quantity', 'Units_Sold']),
        _find_col(df, ['category', 'Category', 'product', 'item'], required=False),
        _find_col(df, ['location', 'Location', 'store_type', 'region'], required=False),
    )


def _load_ecommerce_multi(raw_dir):
    p = _find_file(raw_dir, ['ecommerce_data.csv', 'multi_ecommerce.csv', 'ecommerce.csv'])
    df = pd.read_csv(p, encoding='latin-1')
    return _to_standard(df,
        _find_col(df, ['order_date', 'date', 'Date', 'purchase_date', 'transaction_date']),
        _find_col(df, ['sales', 'Sales', 'revenue', 'amount', 'price']),
        _find_col(df, ['category', 'Category', 'product_category'], required=False),
        _find_col(df, ['country', 'Country', 'region', 'Region'], required=False),
        profit_col=_find_col(df, ['profit', 'Profit', 'margin'], required=False),
    )


def _load_grocery_supermart(raw_dir):
    p = _find_file(raw_dir, [
        'Supermart Grocery Sales - Retail Analytics Dataset.csv',
        'supermart.csv', 'grocery_sales.csv'
    ])
    df = pd.read_csv(p, encoding='latin-1')
    return _to_standard(df,
        _find_col(df, ['order date', 'Order Date', 'date', 'Date']),
        _find_col(df, ['sales', 'Sales']),
        _find_col(df, ['category', 'Category'], required=False),
        _find_col(df, ['region', 'Region', 'city', 'City', 'state'], required=False),
        profit_col=_find_col(df, ['profit', 'Profit'], required=False),
    )


def _load_synthetic_retail(raw_dir):
    sales_p = _find_file(raw_dir, ['sales_transactions.csv', 'transactions.csv', 'sales.csv'])
    df = pd.read_csv(sales_p, encoding='latin-1', nrows=500_000)
    if (raw_dir / 'stores.csv').exists():
        stores   = pd.read_csv(raw_dir / 'stores.csv')
        store_key = _find_col(stores, ['store_id', 'StoreID', 'store'], required=False)
        df_key    = _find_col(df,     ['store_id', 'StoreID', 'store'], required=False)
        if store_key and df_key:
            df = df.merge(stores, left_on=df_key, right_on=store_key, how='left')
    if (raw_dir / 'products.csv').exists():
        prods    = pd.read_csv(raw_dir / 'products.csv')
        prod_key  = _find_col(prods, ['product_id', 'sku_id', 'sku'], required=False)
        df_key    = _find_col(df,    ['product_id', 'sku_id', 'sku'], required=False)
        if prod_key and df_key:
            df = df.merge(prods, left_on=df_key, right_on=prod_key, how='left')
    return _to_standard(df,
        _find_col(df, ['transaction_date', 'date', 'Date', 'order_date', 'sale_date']),
        _find_col(df, ['total_amount', 'amount', 'sales', 'revenue', 'price', 'quantity']),
        _find_col(df, ['category', 'department', 'product_category'], required=False),
        _find_col(df, ['region', 'store_location', 'city', 'state'], required=False),
    )


def _load_ecommerce_performance(raw_dir):
    p = _find_file(raw_dir, ['ecommerce_sales.csv', 'Ecommerce-Sales-Transaction-Data.csv'])
    df = pd.read_csv(p, encoding='latin-1')
    return _to_standard(df,
        _find_col(df, ['date', 'Date', 'order_date', 'transaction_date']),
        _find_col(df, ['sales', 'Sales', 'amount', 'revenue']),
        _find_col(df, ['category', 'Category', 'product'], required=False),
        _find_col(df, ['region', 'Region', 'country', 'location'], required=False),
        profit_col=_find_col(df, ['profit', 'Profit'], required=False),
    )


def _load_africa_nigeria(raw_dir):
    p = _find_file(raw_dir, ['nigeria_retail.csv', 'africa_nigeria.csv', 'africa_synth.csv'])
    df = pd.read_csv(p, encoding='latin-1')
    return _to_standard(df,
        _find_col(df, ['date', 'Date', 'order_date', 'transaction_date']),
        _find_col(df, ['sales', 'Sales', 'amount', 'revenue', 'price']),
        _find_col(df, ['category', 'Category', 'product_category'], required=False),
        _find_col(df, ['state', 'State', 'city', 'City', 'region'], required=False),
    )


def _load_b2b_crm(raw_dir):
    p = _find_file(raw_dir, ['sales_pipeline.csv', 'crm_sales.csv', 'b2b_pipeline.csv'])
    df = pd.read_csv(p, encoding='latin-1')
    if 'deal_stage' in df.columns:
        df = df[df['deal_stage'].str.lower().str.contains('won|closed', na=False)]
    elif 'stage' in df.columns:
        df = df[df['stage'].str.lower().str.contains('won|closed', na=False)]
    return _to_standard(df,
        _find_col(df, ['close_date', 'Close Date', 'date_closed', 'created_date', 'date']),
        _find_col(df, ['close_value', 'deal_value', 'revenue', 'amount', 'sales', 'contract_value']),
        _find_col(df, ['product', 'Product', 'product_name', 'sector'], required=False),
        _find_col(df, ['regional_office', 'region', 'office', 'territory'], required=False),
    )


def _load_kenyan_fmcg(raw_dir):
    p = _find_file(raw_dir, ['kenyan_fmcg.csv', 'kenya_fmcg.csv', 'kenyan_retail.csv'])
    df = pd.read_csv(p, encoding='latin-1')
    return _to_standard(df,
        _find_col(df, ['date', 'Date', 'transaction_date', 'purchase_date']),
        _find_col(df, ['total_amount', 'amount', 'sales', 'revenue', 'price', 'Total Amount']),
        _find_col(df, ['category', 'Category', 'product_category', 'product'], required=False),
        _find_col(df, ['branch', 'Branch', 'city', 'City', 'location', 'store'], required=False),
        customer_col=_find_col(df, ['customer_id', 'Customer ID', 'member_id'], required=False),
    )


def _load_east_africa(raw_dir):
    p = _find_file(raw_dir, ['east_africa_superstore.csv', 'east_africa.csv', 'ea_superstore.csv'])
    df = pd.read_csv(p, encoding='latin-1')
    return _to_standard(df,
        _find_col(df, ['order date', 'Order Date', 'date', 'Date']),
        _find_col(df, ['sales', 'Sales', 'amount', 'revenue']),
        _find_col(df, ['category', 'Category', 'product'], required=False),
        _find_col(df, ['country', 'Country', 'region', 'Region', 'state', 'city'], required=False),
        profit_col=_find_col(df, ['profit', 'Profit'], required=False),
    )


print('All 12 loaders defined.')
```

## Step 5: Main Entry Point

```python
# Map loader names to functions
_LOADERS = {
    '_load_superstore':           _load_superstore,
    '_load_fmcg_profit':          _load_fmcg_profit,
    '_load_retail_store':         _load_retail_store,
    '_load_fmcg_demand':          _load_fmcg_demand,
    '_load_ecommerce_multi':      _load_ecommerce_multi,
    '_load_grocery_supermart':    _load_grocery_supermart,
    '_load_synthetic_retail':     _load_synthetic_retail,
    '_load_ecommerce_performance':_load_ecommerce_performance,
    '_load_africa_nigeria':       _load_africa_nigeria,
    '_load_b2b_crm':              _load_b2b_crm,
    '_load_kenyan_fmcg':          _load_kenyan_fmcg,
    '_load_east_africa':          _load_east_africa,
}


def load_dataset(key, raw_dir=None):
    """
    Load any of the 12 supported datasets.

    Parameters:
        key     : dataset key (see DATASETS dict or run list_datasets())
        raw_dir : path to raw data folder (default: SalesForecast/data/raw/)

    Returns:
        monthly     : pd.Series   — monthly total sales
        cat_monthly : pd.DataFrame — monthly sales by category
        df_raw      : pd.DataFrame — cleaned raw data
    """
    if key not in DATASETS:
        raise ValueError(f"Unknown key '{key}'. Run list_datasets() to see options.")

    if raw_dir is None:
        raw_dir = BASE / 'data' / 'raw'

    meta   = DATASETS[key]
    loader = _LOADERS[meta['loader']]

    print(f"Loading: {meta['name']} ({meta['rows']} rows)")
    df_raw = loader(Path(raw_dir))

    print(f"  Cleaned rows:  {len(df_raw):,}")
    print(f"  Date range:    {df_raw['Order Date'].min().date()} → {df_raw['Order Date'].max().date()}")
    print(f"  Categories:    {list(df_raw['Category'].unique()[:5])}")
    print(f"  Total sales:   ${df_raw['Sales'].sum():,.0f}")

    monthly, cat_monthly = _aggregate(df_raw)
    print(f"  Monthly points: {len(monthly)}")
    return monthly, cat_monthly, df_raw


def list_datasets():
    """Print a table of all available datasets."""
    print(f"\n{'='*72}")
    print(f"{'KEY':<22} {'NAME':<35} {'ROWS':<12} {'INDUSTRY'}")
    print(f"{'='*72}")
    for key, meta in DATASETS.items():
        print(f"{key:<22} {meta['name']:<35} {meta['rows']:<12} {meta['industry']}")
    print(f"{'='*72}")
    print("\nUsage:  monthly, cat_monthly, df = load_dataset('key_name')")
    print("Files:  drop CSVs into SalesForecast/data/raw/")


print('load_dataset() and list_datasets() ready.')
```
