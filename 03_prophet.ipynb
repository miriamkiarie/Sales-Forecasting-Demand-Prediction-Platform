```python
import warnings
warnings.filterwarnings('ignore')
from pathlib import Path
import json

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_squared_error
import joblib

plt.rcParams['figure.dpi'] = 120

BASE      = Path(r'C:\Users\Administrator\SalesForecast')
PROC_DIR  = BASE / 'data' / 'processed'
MODEL_DIR = BASE / 'models'
FIG_DIR   = BASE / 'reports' / 'figures'

print('Libraries loaded.')
```

## 1. Load & Format for Prophet

Prophet requires columns named exactly `ds` (date) and `y` (value).

```python
monthly = pd.read_csv(PROC_DIR / 'monthly_sales.csv', index_col=0, parse_dates=True)
monthly = monthly.squeeze()

# Prophet format
prophet_df = pd.DataFrame({'ds': monthly.index, 'y': monthly.values})

TEST_MONTHS = 12
train_df = prophet_df[:-TEST_MONTHS].copy()
test_df  = prophet_df[-TEST_MONTHS:].copy()

print(f'Prophet training data: {len(train_df)} rows')
prophet_df.head()
```

## 2. Fit Prophet Model

```python
model = Prophet(
    yearly_seasonality  = True,
    weekly_seasonality  = False,   # monthly data — no weekly pattern
    daily_seasonality   = False,
    seasonality_mode    = 'multiplicative',  # better for growing series
    changepoint_prior_scale = 0.15,          # flexibility of trend changes
    interval_width      = 0.95,              # 95% confidence interval
)

# Add US retail holidays — drives spikes in Nov/Dec
model.add_country_holidays(country_name='US')

model.fit(train_df)
print('Prophet model fitted.')
```

## 3. Forecast Test Period & Evaluate

```python
# Create future dataframe for test period
future_test = model.make_future_dataframe(periods=TEST_MONTHS, freq='MS')
forecast    = model.predict(future_test)

# Extract test predictions
test_forecast = forecast.tail(TEST_MONTHS)
y_pred = test_forecast['yhat'].values
y_true = test_df['y'].values

def mape(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

rmse = np.sqrt(mean_squared_error(y_true, y_pred))
mae  = mean_absolute_error(y_true, y_pred)
map_ = mape(y_true, y_pred)

print('=== Prophet Test Metrics ===')
print(f'  MAPE: {map_:.2f}%')
print(f'  RMSE: ${rmse:,.0f}')
print(f'  MAE:  ${mae:,.0f}')
```

## 4. Plot Forecast vs Actual

```python
fig, ax = plt.subplots(figsize=(13, 5))

# Historical
ax.plot(train_df['ds'], train_df['y'], color='#2E75B6', lw=2, label='Training data')
# Actual test
ax.plot(test_df['ds'], test_df['y'], color='#2E75B6', lw=2, linestyle='--', label='Actual (test)')
# Prophet forecast
ax.plot(test_forecast['ds'], test_forecast['yhat'], color='#55A868', lw=2.5, label='Prophet Forecast')
# Confidence band
ax.fill_between(test_forecast['ds'],
                test_forecast['yhat_lower'],
                test_forecast['yhat_upper'],
                alpha=0.20, color='#55A868', label='95% CI')

ax.axvline(test_df['ds'].iloc[0], color='gray', linestyle=':', alpha=0.7, label='Forecast start')
ax.set(title=f'Prophet Forecast — MAPE: {map_:.1f}% | RMSE: ${rmse:,.0f}',
       xlabel='Date', ylabel='Monthly Sales ($)')
ax.legend()
plt.tight_layout()
plt.savefig(FIG_DIR / '08_prophet_forecast.png', bbox_inches='tight')
plt.show()
```

## 5. Prophet Components Plot

This is what makes Prophet powerful — it shows you EXACTLY what is driving the forecast.

```python
fig = model.plot_components(forecast)
plt.suptitle('Prophet Components: Trend + Seasonality + Holidays', fontsize=13, y=1.02)
plt.tight_layout()
plt.savefig(FIG_DIR / '09_prophet_components.png', bbox_inches='tight')
plt.show()
print('Key insight: trend, yearly seasonality, and holiday effects are all separated automatically.')
```

## 6. Future Forecast — 12 Months Ahead

```python
# Refit on ALL data
model_full = Prophet(
    yearly_seasonality      = True,
    weekly_seasonality      = False,
    daily_seasonality       = False,
    seasonality_mode        = 'multiplicative',
    changepoint_prior_scale = 0.15,
    interval_width          = 0.95,
)
model_full.add_country_holidays(country_name='US')
model_full.fit(prophet_df)

# Forecast 12 months ahead
future_12 = model_full.make_future_dataframe(periods=12, freq='MS')
forecast_12 = model_full.predict(future_12)
future_only = forecast_12.tail(12)

fig, ax = plt.subplots(figsize=(13, 5))
ax.plot(prophet_df['ds'], prophet_df['y'], color='#2E75B6', lw=2, label='Historical')
ax.plot(future_only['ds'], future_only['yhat'], color='#55A868', lw=2.5,
        linestyle='--', label='12-month Forecast')
ax.fill_between(future_only['ds'],
                future_only['yhat_lower'],
                future_only['yhat_upper'],
                alpha=0.20, color='#55A868', label='95% CI')
ax.axvline(prophet_df['ds'].iloc[-1], color='gray', linestyle=':', alpha=0.7)
ax.set(title='Prophet — 12-Month Future Forecast', xlabel='Date', ylabel='Sales ($)')
ax.legend()
plt.tight_layout()
plt.savefig(FIG_DIR / '10_prophet_future.png', bbox_inches='tight')
plt.show()
```

## 7. Save Model & Results

```python
joblib.dump(model_full, MODEL_DIR / 'prophet_model.pkl')

prophet_metrics = {'model':'Prophet','mape':round(map_,2),'rmse':round(rmse,0),'mae':round(mae,0)}
with open(MODEL_DIR / 'prophet_metrics.json','w') as f:
    json.dump(prophet_metrics, f)

# Save forecast for dashboard
forecast_12[['ds','yhat','yhat_lower','yhat_upper']].to_csv(
    PROC_DIR / 'prophet_forecast.csv', index=False
)

print('Saved: models/prophet_model.pkl')
print('Saved: data/processed/prophet_forecast.csv')
print('Next: open 04_ML_forecasting.ipynb')
```
